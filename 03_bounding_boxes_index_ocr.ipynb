```python
import pytesseract
import cv2
```

```python
image = cv2.imread("dataset/dataset/training_data/images/00040534.png")
base_image = image.copy()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imwrite("temp/index_gray.png", gray)
```

```python
blur = cv2.GaussianBlur(gray, (7,7), 0)
```

```python
cv2.imwrite("temp/index_blur.png", blur)
```

```python
thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)
```

```python
cv2.imwrite("temp/index_thresh.png", thresh)
```

```python
kernal = cv2.getStructuringElement(cv2.MORPH_RECT, (25, 1))
```

```python
cv2.imwrite("temp/index_kernal.png", kernal)
```

```python
dilate = cv2.dilate(thresh, kernal, iterations=1)
```

```python
cv2.imwrite("temp/index_dilate.png", dilate)
```

```python
def sort_contours(cnts):
    boundingBoxes = [cv2.boundingRect(c) for c in cnts]
    (cnts, boundingBoxes) = zip(*sorted(zip(cnts, boundingBoxes), key=lambda b: b[1][1]))
    return cnts
```

```python
cnts = sort_contours(cnts)
```

```python
cnts, hierarchy = cv2.findContours(dilate, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
cnts = cnts[0] if len(cnts) == 2 else cnts[1]
```

```python
cnts = sorted(cnts, key=lambda x: cv2.boundingRect(x)[0])
```

```python
for c in cnts:
    x, y, w, h = cv2.boundingRect(c)
    if h > 10 and w > 20:
        roi = image[y:y+h, x:x+h]
        cv2.imwrite("temp/index_roi.png", roi)
        cv2.rectangle(image, (x, y), (x+w, y+h), (36, 255, 12), 2)
cv2.imwrite("temp/index_bbox_new.png", image)
```
